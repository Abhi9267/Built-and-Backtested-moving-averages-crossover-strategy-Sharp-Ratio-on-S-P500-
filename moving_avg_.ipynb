```python
import numpy as np
import pandas as pd
data= np.random.rand(10)
#print(prices)
prices=pd.DataFrame(data)
print(prices)

short = prices.rolling(3).mean()
long  = prices.rolling(5).mean()
#print(short)
#print(long)

# Signal: 1 if short MA above long MA else 0
signal = (short > long).astype(int)
#print(signal)

# Shift signal so you trade on the NEXT day (avoid lookahead bias)
position = signal.shift(1)
#print(position)
# Daily returns
returns = prices.pct_change()
#print(returns)

# Strategy returns = position * returns
strategy_returns = position * returns

# Sharpe
annual_sharpe = (strategy_returns.mean() / strategy_returns.std()) * (252 ** 0.5)
equity = (1 + strategy_returns).cumprod()
print(equity)
equity = (1 + strategy_returns).cumprod()
rolling_max = equity.cummax()
print(rolling_max)

drawdown = (equity - rolling_max) / rolling_max

print(drawdown)
#print("Sharpe:", sharpe)

#print(strategy_returns)
```
